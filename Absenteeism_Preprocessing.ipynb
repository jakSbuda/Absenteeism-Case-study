```python
new_df = new_df.drop(['Reason for Absence'], axis=1)
new_df
```

```python
# 1 Various diseases 
# 2 pregnancy and giving birth
# 3 poisoning
# 4 light diseases
reason_type_1 = reason_col.loc[:,1:14].max(axis=1)
reason_type_2 = reason_col.loc[:,15:17].max(axis=1)
reason_type_3 = reason_col.loc[:,18:21].max(axis=1)
reason_type_4 = reason_col.loc[:,22:].max(axis=1)
```

```python
# combine new_df with reason for absent
new_df = pd.concat([new_df, reason_type_1,reason_type_2,reason_type_3,reason_type_4], axis = 1)
new_df
```

```python
# change column names 
new_df.columns.values
```

```python
col_names = ['Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours', 'Reason_1', 'Reason_2', 'Reason_3', 'Reason_4']
```

```python
new_df.columns = col_names
```

```python
new_df.head()
```

```python
# Reorder columns in df
col_names_sort = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 'Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']
```

```python
new_df = new_df[col_names_sort]
new_df.head()
```

```python
# Create check point
df_reason = new_df.copy()
```

```python
#Analysis of date coloumn, convert to timestamp
df_reason['Date']
```

```python
df_reason['Date'] = pd.to_datetime(df_reason['Date'],format='%d/%m/%Y')
```

```python
df_reason.shape[0]
```

```python
# Extract month value
list_month = []

for i in range(df_reason.shape[0]):
  list_month.append(df_reason["Date"][i].month) 

df_reason['Month Value'] = list_month

df_reason.head()
```

```python
# Extract day of the week value
def day_value(date_value):
  return date_value.weekday()
```

```python
df_reason['Day of the week'] = df_reason['Date'].apply(day_value)
```

```python
df_reason.head()
```

Transport expense, distance to work and age, children and pets dont need any transformation  

```python
# Education transformation 
df_reason['Education'].unique()

# 1 high school
# 2 graduate
# 3 postgraduate
# 4 master's or Phd
```

Combine 2, 3 and 4 into one variable as they are all graduates

```python
df_reason['Education'] = df_reason['Education'].map({1:0, 2:1 , 3:1, 4:1})
```
